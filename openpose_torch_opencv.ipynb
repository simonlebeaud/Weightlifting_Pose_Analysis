# OpenPose

This code is taken from the repository [Hzzone/pytorch-openpose](https://github.com/Hzzone/pytorch-openpose). The original repository from openpose and the library associated is no longer maintained and this torch implemantation is perfect to be used as so.  

OpenPose introduced a multi-person pose estimation framework that simultaneously detects and localizes human body key points in images or videos. It employs a multi-stage CNN architecture to estimate human poses from images or videos.  


#### Pros of OpenPose

* Free to use: OpenPose is entirely free, making it an attractive option for budget-conscious users.
* Stable performance: OpenPose delivers reliable performance in tracking and estimating human poses in various contexts.


#### Cons of OpenPose

* High hardware requirements: To run OpenPose smoothly, high-tier computer specifications and a powerful NVIDIA graphics card are necessary. This can pose difficulties for users with budget or outdated systems.
* Limitations in 3D file generation: OpenPose only supports 2D annotations and does not provide direct 3D file generation, limiting its application in certain contexts.
* The model is old and new model are much faster and performant than openpose.


```python
import cv2
import matplotlib.pyplot as plt
import copy
import numpy as np

from src_openpose import model
from src_openpose import util
from src_openpose.body import Body

body_estimation = Body('models/body_pose_model.pth')

test_image = 'images/demo.jpg'
oriImg = cv2.imread(test_image)  # B,G,R order
candidate, subset = body_estimation(oriImg)
canvas = copy.deepcopy(oriImg)
canvas = util.draw_bodypose(canvas, candidate, subset)


plt.imshow(canvas[:, :, [2, 1, 0]])
plt.axis('off')
plt.show()
```
